## 介绍NumPy的使用

1. 基础的加减乘法

````python
import numpy as np
x = np.array([1,2,3])
y = np.array([4,5,6])
print(x + y)
print(y-x)
print(x*y)
````

#### 多维数组

1. 一维数组是响亮，二维数组是矩阵，三维数组以上就是张量
2. 标量当作向量来计算，叫广播

````python
xx = np.array([[1,2],[4,5]])
print(xx.shape)
print(xx.dtype)
print(xx.ndim)
print(xx.size)
print(xx.itemsize)
print(xx.nbytes)
````

````python
yy = np.array([[2,3],[5,6]])
print(xx + yy)
print(xx*yy)
print(yy*3) 
print(yy[1])
print(yy[1][1])
for i in yy:
    print(i)

yy_flat = yy.flatten()
print(yy_flat)
yy_search = yy_flat[np.array([1,2])]
print(yy_search)
yy_flat[yy_flat > 2]

````

````python
import matplotlib.pyplot as plt

x = np.arange(0,6,0.1)
y = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y,label="sin")
plt.plot(x,y2,label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.show()
````
